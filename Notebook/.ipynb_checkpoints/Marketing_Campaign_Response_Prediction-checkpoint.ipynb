This assignment provides an opportunity to build an ANN model using the a bank marketing dataset

https://www.kaggle.com/sriharipramod/bank-loan-classification

Target variable - personal loan - reflects whether a customer accepted a personal loan offer.

Load the data and perform EDA.

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
import seaborn as sns
```

```python
#Reading the data

bank = pd.read_csv('UniversalBank.csv')
bank.info()
```

1. Evaluate missing values.

```python
# Explore missing values

sns.heatmap(bank.isnull(), cbar=False)
```

2. Assess target class distribution.

```python
# Exploratory analysis
# Evaluate target variable distribution

bank['Personal Loan'].mean()
```

3. Pre-process data as needed.
4. Assess information value of individual features (correlation analysis and pairlot).

5 pts.

```python
bank.corr()['Personal Loan'].sort_values(ascending=False)
```

```python
# Pair Plot

sns.pairplot(bank, hue = 'Personal Loan')
```

```python
# drop age (Because the correlation with experience is high.)

bank = bank.drop(['Age'],axis=1)
```

```python
# Dropping ID & ZIP Code, not necessarily correlate to a customer accepting a loan
bank.drop(["ID","ZIP Code"], axis=1, inplace=True)

cat_cols = ["Personal Loan", "Securities Account",
           "Online","CreditCard","CD Account"]
bank = pd.get_dummies(bank,columns=cat_cols,drop_first=True,)

bank['Education'] = bank['Education'].astype('category',copy=False)
bank['Family'] = bank['Family'].astype('category',copy=False)
bank['Experience'] = bank['Experience'].astype('category',copy=False)
```

```python
bank.corr()['Personal Loan_1'].sort_values(ascending=False)
```

```python
#Splitting
X = pd.DataFrame(bank)
X = X.drop(['Personal Loan_1'],axis=1)
y = pd.DataFrame(bank)
y = y[['Personal Loan_1']]
```

```python
#Rescale data for analysis
scaler=MinMaxScaler()

X_=scaler.fit_transform(X)

X_rescaled=pd.DataFrame(X_,columns=X.columns)
```

```python
X_rescaled.describe().transpose()
```

5. Split the data 70/30 into training and test datasets.

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.30,random_state=1)

scaler = MinMaxScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

6. Develop an ANN model (MLPClassifier) with a single hidden layer with 20 nodes.

```python
MLPC_model = MLPClassifier(hidden_layer_sizes=20, activation='logistic', solver='adam', random_state=1, max_iter=1000)

MLPC_model.fit(X_train, y_train.values.ravel())

y_pred_mlpc = MLPC_model.predict(X_test)
```

7. Assess model performance, provide the confusion matrix, classification report and ROC AUC values.


```python
print(confusion_matrix(y_test,y_pred_mlpc))

print(classification_report(y_test,y_pred_mlpc))
```

>There are 6 false positives out of the 123 predictions of personal loans being accepted. Out of the 123 predictions, 117 were accurate. Implying the model has good discriminitive abilities.

>In general, the scores provides by the classification report indicates the model is a good fit.

```python
print('ROC AUC: ', roc_auc_score(y_test,y_pred_mlpc))
```

>.90 ROC AUC score, the curve sugests outstanding discrimination, which is good. 

8. Assess the effect of the number of hidden nodes (5-30) on model accuracy.
9. Identify optimal number of nodes.

5 pts.

```python
from sklearn.model_selection import cross_val_score
min_hidden_layer_size = 5
max_hidden_layer_size = 30
cv_scores = [ ]

for s in range(min_hidden_layer_size,max_hidden_layer_size,5):
    MLPC_model = MLPClassifier(hidden_layer_sizes=s, activation='logistic', solver='lbfgs', random_state=1, max_iter=10000)
    scores = cross_val_score(MLPC_model,X_train,y_train.values.ravel(),cv = 5,scoring = "accuracy")
    cv_scores.append(scores.mean())
```

```python
sns.lineplot(x=range(min_hidden_layer_size,max_hidden_layer_size,5), y=cv_scores)
```

> The optimal number of nodes for this model would be between 15-20 as that is the highest peak in the graph above, which plots the accuracy scores of the model at each possible hidden nodes (between 5 and 30).
