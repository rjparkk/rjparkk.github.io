```
# Pre-load packages

import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import math
import numpy as np
from matplotlib import pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer
```

```
diabetes = pd.read_csv('diabetes.csv')
```

```
X = diabetes[['Pregnancies','Glucose','BloodPressure','Age'
,'SkinThickness','Insulin', 'BMI','DiabetesPedigreeFunction']]
y = diabetes[['Outcome']]
y = pd.get_dummies(y)
```

```
#Create a dataframe from Xandy for exploratory analysis

db_data = pd.concat([X,y],axis=1)
```

```
sns.pairplot(db_data,hue='Outcome')
```

```
# MinMax Scalar transforms features into [0,1]
# Formula: X_rescaled = (Xi - Xmin) / (Xmax-Xmin) 

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_ = scaler.fit_transform(X)

X_rescaled= pd.DataFrame(X_, columns=X.columns)
```

```
X_rescaled.describe()
```

```
X_train, X_test, y_train, y_test = train_test_split(X_rescaled, y, test_size=0.3, random_state=1)
```

5. Develop a kNN model to predict diabetes occurence using k = 5
6. Determine the optimal value of k based on the ROC AUC (Hint: scoring="roc_auc")

10 pts.

```
knn = KNeighborsClassifier(n_neighbors=5, metric='euclidean')
knn.fit(X_train,y_train)

y_pred = knn.predict(X_test)

print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

> b14 false negatives and 32 false positives

> F score is .80, high score.



```
from sklearn.metrics import roc_auc_score
print('Roc AUC', roc_auc_score(y_test,y_pred))
```

> ROC area under the curve is .76, which is an acceptible classifier. 

```
# Determing the optimal value of k based on the ROC AUC

from sklearn.model_selection import cross_val_score
max_K = 100
cv_scores = [ ]

for K in range(1,max_K):
  knn = KNeighborsClassifier(n_neighbors = K)
  scores = cross_val_score(knn,X_train,y_train.values.ravel(),cv = 5, scoring = "accuracy")
  cv_scores.append(scores.mean())
```

```
sns.lineplot(x=range(1,max_K), y=cv_scores)
```

```
cv_scores[:20]
```

> Cross validation is taking the training dataset, splitting it 5 ways. Using four of the subsamples to train, and the fifth time is used to access model performance. We do this cross validation five times, as that is how many splits we can have in terms of training vs test data. 

> The graph shows the cross validation by trying different values of K. The graph goes up to maximum value of K of 100. 

> Looking at the cross validation scores, we see it at the higest when the value of K equals 12, scoring at .73
